```python
class XO:
    
    def __init__(self):
        
        self.rv=[]
        self.lb=[1,2,3,4,5,6,7,8,9]
        self.lb2=[1,2,3,4,5,6,7,8,9]
        self.flag=0
        self.flag2=0
        self.flag3=0
        self.dispboard()
    
    def dispboard(self):
        print(f'-------------------------------')
        print(f'|     {self.lb[0]}    |     {self.lb[1]}     |  {self.lb[2]}   |')
        print("-------------------------------")
        print(f'|     {self.lb[3]}    |     {self.lb[4]}     |  {self.lb[5]}   |')
        print("-------------------------------")
        print(f'|     {self.lb[6]}    |     {self.lb[7]}     |  {self.lb[8]}   |')
        print(f'-------------------------------') 
        self.sGame()
        
    def sGame(self):
        
 
        if self.flag==0:      
            self.x=int(input("O please enter number : " )) 
            if self.x in self.lb:
                self.lb[self.x-1]='o'
                self.flag=1
            else:
                print("O please enter another number")
        elif self.flag==1:
            self.x=int(input("X please enter number : " ))
            if self.x in self.lb:
                self.lb[self.x-1]='x'
                self.flag=0
            else:
                print("X please enter another number")

        if self.flag2==0:
            #row
            if self.lb[0]==self.lb[1]==self.lb[2]=='o' or self.lb[3]==self.lb[4]==self.lb[5]=='o' or self.lb[6]==self.lb[7]==self.lb[8]=='o':
                self.flag2=1
            elif self.lb[0]==self.lb[1]==self.lb[2]=='x' or self.lb[3]==self.lb[4]==self.lb[5]=='x' or self.lb[6]==self.lb[7]==self.lb[8]=='x':
                self.flag2=2
            #Col
            if self.lb[0]==self.lb[3]==self.lb[6]=='o' or self.lb[1]==self.lb[4]==self.lb[7]=='o' or self.lb[2]==self.lb[5]==self.lb[8]=='o':
                self.flag2=1
            elif self.lb[0]==self.lb[3]==self.lb[6]=='x' or self.lb[1]==self.lb[4]==self.lb[7]=='x' or self.lb[2]==self.lb[5]==self.lb[8]=='x':
                self.flag2=2
            #cor
            if self.lb[0]==self.lb[4]==self.lb[8]=='o' or self.lb[2]==self.lb[4]==self.lb[6]=='o':
                self.flag2=1
            elif self.lb[0]==self.lb[4]==self.lb[8]=='x' or self.lb[2]==self.lb[4]==self.lb[6]=='x':
                self.flag2=2
        self.gameSt()
        if self.flag3==0:
            self.dispboard()
        
        
    def gameSt(self):
        if self.flag2==1:
            print("////////////////////////")
            print("////O WiIINNNEEERRRR////")
            print("////////////////////////")
            self.flag2=9
            self.flag3=1
            self.flag=9
            self.dispboard()
        if self.flag2==2:
            print("////////////////////////")
            print("////X WiIINNNEEERRRR////")
            print("////////////////////////")
            self.flag2=9
            self.flag3=1
            self.flag=9
            self.dispboard()
        
        
        return None
    
```
